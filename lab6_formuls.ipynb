
### моя исходная функция

`f(x₁, x₂) = 64x₁² + 126x₁x₂ + 64x₂² - 10x₁ + 30x₂ + 13`

### альтернативная запись в матричной форме

`f(x) = 1/2 * xᵀ A x + bᵀ x + c`

Где:
*   `x` — вектор переменных, `x = [x₁, x₂]ᵀ`.
*   `A` — симметричная матрица коэффициентов при квадратичных членах, `A = [[128, 126], [126, 128]]`.
*   `b` — вектор коэффициентов при линейных членах, `b = [-10, 30]ᵀ`.
*   `c` — свободный член, `c = 13`.

### градиент функции

Градиент функции был вычислен путем нахождения частных производных:
*   Частная производная по `x₁`: `∂f/∂x₁ = 128x₁ + 126x₂ - 10`
*   Частная производная по `x₂`: `∂f/∂x₂ = 126x₁ + 128x₂ + 30`

Таким образом, вектор градиента записывается как:
`f(x₁, x₂) = [128x₁ + 126x₂ - 10, 126x₁ + 128x₂ + 30]ᵀ`

### аналитический минимум

Точка аналитического минимума была найдена путем решения системы уравнений, полученной из условия равенства градиента нулю: `∇f = 0`.

Решив систему:
1. `128x₁ + 126x₂ = 10`
2. `126x₁ + 128x₂ = -30`

Были получены следующие координаты точки минимума:
`(x₁*, x₂*) ≈ (9.96, -10.04)`
